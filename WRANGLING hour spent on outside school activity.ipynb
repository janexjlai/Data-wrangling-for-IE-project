```python
import pandas as pd
import numpy as np
```

```python
# read multi indexes with $header 
files = '49010do021_201204.xls'
sheets = 'Table_11'
data = pd.read_excel(files, sheet_name=sheets, skiprows=5, header = [0,1])
```

```python
data
```

```python
data.columns
```

```python
# only selected data are kept
# selected person data
data_p = data.iloc[15:20,0:-1]
data_p
```

```python
data_p.dtypes
```

```python
data_p.iloc[:,0:2] = data_p.iloc[:,0:2].astype(float)
```

```python
data_p.dtypes
```

```python
data_p.columns = data_p.columns.droplevel(0)
```

```python
data_p.columns
```

```python
data_pp = pd.DataFrame()
data_pp['5-8'] = data_p['5–8']/data_p['5–8'].sum()
data_pp['9–11'] = data_p['9–11']/data_p['9–11'].sum()
data_pp['12–14'] = data_p['12–14']/data_p['12–14'].sum()
```

```python
data_pp.reset_index(inplace=True)
```

```python
data_pp
```

```python
data_pp.rename(columns={"NUMBER ('000)": "Hours in two weeks"},inplace=True)
```

```python
newdata = data_pp.melt(id_vars=['Hours in two weeks'], var_name = 'Age group', value_name = 'Proportion')
```
