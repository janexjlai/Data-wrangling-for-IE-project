```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# method 2: read_excel
# read multi indexes with $header 
files = 'aihw-phe-215-data-tables.xls'
sheets = 'Table S1'
data = pd.read_excel(files, sheet_name=sheets, skiprows=1, header = [0,1])
```

```python
data
```

```python
# only the birth cohurt data should be kept
data = data.iloc[:,0:15]
```

```python
# filter null row
data = data[~data.isnull().all(1)]
data
```

```python
# filter null columns
data.dropna(axis= 1, how= 'all', inplace = True)
data
```

```python
# drop rows with 2-5 years old survey data
data1 = data.drop(labels=['2–5'])
```

```python
data2 = data1.drop(labels=['% (95% CI).1','Birth cohort'], level = 1, axis=1)
```

```python
data2
```

```python
### data for person
data_p = data2.iloc[0:3,:]
data_p
```

```python
data_p = data_p.unstack().reset_index()
data_p
```

```python
data_p.drop(columns = ['Age at survey'], inplace=True)
data_p.rename(columns = {'Persons': 'Age at survey', 0: 'Proportion', 'level_0':'year measured'},inplace=True)
data_p
```

```python
data_p.replace('Most recently born birth cohort\n(measured 2014–15)', '2014-2015', inplace=True)
data_p.replace('Birth cohort born 4 years earlier\n(measured 2011–12)', '2011-2012', inplace=True)
data_p.replace('Birth cohort born 20 years earlier\n(measured 1995)', '1995', inplace=True)
data_p.replace('Birth cohort born 8 years earlier\n(measured 2007–08)', '2007-2008', inplace=True)
```

```python
data_p.replace('6–9', '6-9', inplace=True)
data_p.replace('10–13', '10-13', inplace=True)
data_p.replace('14–17', '14-17', inplace=True)
```

```python
data_p['Gender'] = 'Persons'
```

```python
data_p
```

```python
####### male data
data_m = data2.iloc[3:6,:]
```

```python
data_m = data_m.unstack().reset_index()
data_m.drop(columns = ['Age at survey'], inplace=True)
data_m.rename(columns = {'Persons': 'Age at survey', 0: 'Proportion', 'level_0':'year measured'},inplace=True)


# inplace values to make dataframe more clean
# and avoid encoding issues
data_m.replace('Most recently born birth cohort\n(measured 2014–15)', '2014-2015', inplace=True)
data_m.replace('Birth cohort born 4 years earlier\n(measured 2011–12)', '2011-2012', inplace=True)
data_m.replace('Birth cohort born 20 years earlier\n(measured 1995)', '1995', inplace=True)
data_m.replace('Birth cohort born 8 years earlier\n(measured 2007–08)', '2007-2008', inplace=True)

data_m.replace('6–9', '6-9', inplace=True)
data_m.replace('10–13', '10-13', inplace=True)
data_m.replace('14–17', '14-17', inplace=True)

data_m['Gender'] = 'Boys'

data_m
```

```python
####### female data
data_f = data2.iloc[6:,:]
```

```python
# for female
data_f = data_f.unstack().reset_index()
data_f.drop(columns = ['Age at survey'], inplace=True)
data_f.rename(columns = {'Persons': 'Age at survey', 0: 'Proportion', 'level_0':'year measured'},inplace=True)


# inplace values to make dataframe more clean
# and avoid encoding issues
data_f.replace('Most recently born birth cohort\n(measured 2014–15)', '2014-2015', inplace=True)
data_f.replace('Birth cohort born 4 years earlier\n(measured 2011–12)', '2011-2012', inplace=True)
data_f.replace('Birth cohort born 20 years earlier\n(measured 1995)', '1995', inplace=True)
data_f.replace('Birth cohort born 8 years earlier\n(measured 2007–08)', '2007-2008', inplace=True)

data_f.replace('6–9', '6-9', inplace=True)
data_f.replace('10–13', '10-13', inplace=True)
data_f.replace('14–17', '14-17', inplace=True)

data_f['Gender'] = 'Girls'

data_f
```

```python
# merge dataframes
data_final = pd.concat([data_p, data_f, data_m],axis=0)
data_final
```

```python
# write data
data_final.to_csv('obesity rate by birth cohurt_v1.0.csv', encoding='utf8')
```
