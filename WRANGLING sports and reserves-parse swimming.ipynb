```python
# a new dataframe
new_col=data.columns.tolist()
#new_col.append('MainSportPlayed')
print(new_col)
df = pd.DataFrame(columns=new_col)
df


```

```python
# new dataframe selected sports type
# as long as there is park/reserve in name
for i in range(data.shape[0]):
    
    names = data.iloc[i,:].FacilityName
    sports = data.iloc[i,:].SportsPlayed
    # sports
    # this may deal with null values
    if 'park' in names or 'reserve' in names:
        new_row = {}
        new_row = {'Facility_ID': data.iloc[i,:].Facility_ID,
                   'FacilityName': data.iloc[i,:].FacilityName,
                   'StreetNo': data.iloc[i,:].StreetNo,
                   'StreetType': data.iloc[i,:].StreetType,
                   'SuburbTown': data.iloc[i,:].SuburbTown,
                   'Latitude': data.iloc[i,:].Latitude,
                   'Longitude': data.iloc[i,:].Longitude,
                   'SportsPlayed': 'Park / Reserve',
                   #'MainSportsPlayed': 'Walking',
                   'LGA': data.iloc[i,:].LGA
            
        }
        #print(names, data.iloc[i,:].SportsPlayed)
        df = df.append(new_row, ignore_index=True)
        
    if sports in sports_list:
        df = df.append(data.iloc[i,:], ignore_index=True)

        
    
```

```python
# read in API
with open('google_map_api_key.txt') as f:
    api_key = f.readline()
    f.close
    
#api_key
```

```python
df2 = df[df.Postcode.isnull()]
```

```python
df2.shape
```

```python
# reset the indexes
df2.reset_index(inplace=True, drop=True)
df2
```

```python
# reverse geocoding 

examp = 'https://maps.googleapis.com/maps/api/geocode/json?latlng={},{}&key={}'

for i in range(df2.shape[0]):
    lng = df2.iloc[i,:].Longitude
    lat = df2.iloc[i,:].Latitude
    

    q = examp.format(lat, lng, api_key)
    #print(q)
    res = requests.get(q).json()
    result = res.get('results')
    
    # only for valid results
    if len(result) != 0:
        pp = [p.get('short_name') for p in result[0].get('address_components') if p.get('types') == ['postal_code']]
        #ss = [p.get('short_name') for p in result[0].get('address_components') if 'locality' in p.get('types')]
        postcode = pp[0] if len(pp) > 0 else None
        #sub = ss[0] if len(ss) > 0 else None
        #print(postcode,sub)
        print(postcode)
        
        df2.loc[i,'Postcode'] = postcode  
        #df2.loc[i,'SuburbTown'] = sub
        
        #print(postcode, sub)
        #if df2.iloc[i,:].SuburbTown == np.nan:
            #print('yes')
            #df2.iloc[i,:].SuburbTown = [p.get('short_name') for p in result[0].get('address_components') if 'locality' in p.get('types')][0]

        
    # uncomment for tesing
    #if i == 8:
        #break
print('done')
```
