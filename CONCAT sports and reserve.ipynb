```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('sports and reserve_v1.1.csv')
data.head()
```

```python
data.StreetType = data.StreetType.str.capitalize()
```

```python
data.StreetType.unique()
```

```python
# keep naming convention
# use full name instead of short name
data.StreetType.replace('Hwy', 'Highway', inplace=True)
data.StreetType.replace('Blvd', 'Boulevard', inplace=True)
data.StreetType.replace('Strret', 'Street', inplace=True)
data.StreetType.replace('Street\xa0\xa0', 'Street', inplace=True)
data.StreetType.replace('Steet', 'Street', inplace=True)
data.StreetType.replace('.', np.nan , inplace=True)
```

```python
data.StreetType.unique()
```

```python
data.Postcode = data.Postcode.astype(int)
```

```python
data2 = data.copy()
data2.head()
```

```python
state = 'VIC'
    
for i in range(data2.shape[0]):
    num = data2.iloc[i,:].StreetNo
    names = data2.iloc[i,:].StreetName
    types = data2.iloc[i,:].StreetType
    sub = data2.iloc[i,:].Suburb
    post = data2.iloc[i,:].Postcode
    
    # add1: first part of address
    add1 = []
    
    add1.append(num if str(num) != 'nan' else '')
    add1.append(names if str(names) != 'nan' else '')
    add1.append(types if str(types) != 'nan' else '')
    
    # add2: second part of address
    add2 = []
    
    add2.append(sub)
    add2.append(state)
    add2.append(post)
    
    
    formatted1 = ' '.join(p for p in add1 if p != '')
    formatted2 = ' '.join(str(p) for p in add2)
    add_f = ', '.join([formatted1,formatted2, 'Australia'])
    #print(add_f)
    
    data2.loc[i,'StreetNo'] = add_f
    
    # for testing purpose
    # uncomment to run the whole loop
    #if i == 30:
        #break
```

```python
data2[data2.StreetNo.isnull()]
```

```python
data3 = data2.copy()
```

```python
data3.drop(columns=['StreetName', 'StreetType', 'LGA'], inplace = True)
data3.rename(columns={'StreetNo': 'formatted_address'}, inplace = True)
```

```python
# add for keeping naming convention
data3.rename(columns = {'Facility_ID':'place_id', 'FacilityName': 'name', 'Type':'type'}, inplace = True)
```

```python
# write data
data3.to_csv('sports and reserve_v2.1.csv', encoding = 'utf8')
```
