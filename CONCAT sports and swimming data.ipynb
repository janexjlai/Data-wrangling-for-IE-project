```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('sports and reserve_v1.1.csv')
data.head()
```

```python
data.rename(columns = {'Facility_ID':'place_id', 'FacilityName': 'name', 'Type':'type'}, inplace = True)
```

```python
swim = pd.read_csv('sports and reserve swimming only_v1.0.csv')
swim.head()
```

```python
swim.drop(columns=['Unnamed: 0', 'LGA'], inplace = True)
swim.rename(columns = {'FacilityName': 'name', 'Type':'type'}, inplace = True)
data.drop(columns=['LGA'], inplace = True)
```

```python
swim.head()
```

```python
data.head()
```

```python
df = pd.concat([data, swim])
df.shape
```

```python
df.head()
```

```python
df_type = pd.concat([df.place_id, df.type], axis= 1)
df_type.head()
```

```python
sportstype = df_type.groupby('place_id')['type'].apply(','.join).reset_index()
```

```python
for i in sportstype.type:
    if 'Swimming' in i:
        print(i)
```

```python
sportstype.head()
```

```python
sportstype.shape
```

```python
df.drop_duplicates(subset='place_id',inplace=True)
df.shape
```

```python
df1 = df.copy()
df1.head()
```

```python
df1.drop(columns=['type'], inplace=True)
```

```python
new = pd.merge(df1,sportstype, on = 'place_id')
new.head()
```

```python
############## format the address
```

```python
new.StreetType = new.StreetType.str.capitalize()
```

```python
new.StreetType.unique()
```

```python
# keep naming convention
# use full name instead of short name
new.StreetType.replace('Hwy', 'Highway', inplace=True)
new.StreetType.replace('Blvd', 'Boulevard', inplace=True)
new.StreetType.replace('Strret', 'Street', inplace=True)
new.StreetType.replace('Street\xa0\xa0', 'Street', inplace=True)
new.StreetType.replace('Steet', 'Street', inplace=True)
new.StreetType.replace('.', np.nan , inplace=True)
```

```python
new.Postcode = new.Postcode.astype(int)
```

```python
data2 = new.copy()
data2.head()
```

```python
state = 'VIC'
    
for i in range(data2.shape[0]):
    num = data2.iloc[i,:].StreetNo
    names = data2.iloc[i,:].StreetName
    types = data2.iloc[i,:].StreetType
    sub = data2.iloc[i,:].Suburb
    post = data2.iloc[i,:].Postcode
    
    # add1: first part of address
    add1 = []
    
    add1.append(num if str(num) != 'nan' else '')
    add1.append(names if str(names) != 'nan' else '')
    add1.append(types if str(types) != 'nan' else '')
    
    # add2: second part of address
    add2 = []
    
    add2.append(sub)
    add2.append(state)
    add2.append(post)
    
    
    formatted1 = ' '.join(p for p in add1 if p != '')
    formatted2 = ' '.join(str(p) for p in add2)
    if formatted1 == '':
        add_f = ', '.join([formatted2, 'Australia'])
    else:
        add_f = ', '.join([formatted1, formatted2, 'Australia'])
    #print(add_f)
    
    data2.loc[i,'StreetNo'] = add_f
    
    # for testing purpose
    # uncomment to run the whole loop
    #if i == 30:
        #break
```

```python
data2[data2.StreetNo.isnull()]
```

```python
data3 = data2.copy()
data3.head()
```

```python
data3.drop(columns=['StreetName', 'StreetType'], inplace = True)
data3.rename(columns={'StreetNo': 'formatted_address'}, inplace = True)
```
