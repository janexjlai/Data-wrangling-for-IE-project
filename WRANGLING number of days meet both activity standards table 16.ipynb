```python
# include gender
import pandas as pd
import numpy as np
```

```python
# read multi indexes with $header 
files = 'Table 16 Number of days met physical and screen-based activity recommendations by sex and age, Children aged 2-17 years.xls'
sheets = 'Table_16_1'
data = pd.read_excel(files, sheet_name=sheets, skiprows=4, header = [2,2], index_col = [0,1])
```

```python
# data inspection
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
# level 1 indexing is unnecessary
# drop the level 1 indexing for columns
data.columns = data.columns.droplevel(1)
```

```python
# there are clearly many null rows and columns in this dataframe
# delete null rows
# keep a copy data1
data1 = data[~data.isnull().all(1)]
data1
```

```python
# drop null columns, keep in data1
data1.dropna(axis= 1, how= 'all', inplace = True)
data1
```

```python
# we omit data for kids aged from 2 to 4
data2 = data1.iloc[:,1:6]
```

```python
# Number of days over 7 days met screen-based activity recommendations
# data_both
data_both = data2.loc['Number of days over 7 days met both physical and screen-based activity recommendations(a)(b)(c)']
data_both.index.names = ['Number of days per week']
```

```python
data_both
```

```python
##### data for male
data_both_m = data_both.iloc[0:6,:]
data_both_m
```

```python
# make a copy
data3 = data_both_m.copy()
data3.loc['2-7 days'] = data3.iloc[1:5,:].sum()
data3.loc['less than 1 day'] = data3.loc['Total(d)'].sub(data3.loc['2-7 days'])
```

```python
# reset index
data3.reset_index(inplace=True)
```

```python
# to avoid encoding issues
# change index values
data3.replace('1–2 days', '1-2 days', inplace=True)
data3.replace('3–4 days', '3-4 days', inplace=True)
data3.replace('5–6 days', '5-6 days', inplace=True)
data3.replace('Total(d)', 'Total', inplace=True)
```

```python
data3
```

```python
data3 = data3[~(data3['Number of days per week'] == 'Total')]
data3 = data3[~(data3['Number of days per week'] == '2-7 days')]
data3 = data3[~(data3['Number of days per week'] == 'None')]
data3
```

```python
# melt dataframe
data3_new = data3.melt(id_vars=['Number of days per week'], var_name='Age Group', value_name='Number of kids(\'000)')
```

```python
# avoid encoding issues
data3_new.replace('5–8','5-8', inplace=True)
data3_new.replace('9–11','9-11', inplace=True)
data3_new.replace('12–14','12-14', inplace=True)
data3_new.replace('15–17','15-17', inplace=True)
data3_new.replace('Total 5–17 years', 'Total 5-17 years', inplace=True)
```

```python
# for gender male
data3_new['Gender'] = 'Male'
data3_new
```

```python
#### for female data
data_both_f = data_both.iloc[6:12,:]
data_both_f
```

```python
# same operations applied to female data
# make a copy
data4 = data_both_f.copy()
data4.loc['2-7 days'] = data4.iloc[1:5,:].sum()
data4.loc['less than 1 day'] = data4.loc['Total(d)'].sub(data4.loc['2-7 days'])

# reset index
data4.reset_index(inplace=True)

# to avoid encoding issues
# change index values
data4.replace('1–2 days', '1-2 days', inplace=True)
data4.replace('3–4 days', '3-4 days', inplace=True)
data4.replace('5–6 days', '5-6 days', inplace=True)
data4.replace('Total(d)', 'Total', inplace=True)

# filter data
data4 = data4[~(data4['Number of days per week'] == 'Total')]
data4 = data4[~(data4['Number of days per week'] == '2-7 days')]
data4 = data4[~(data4['Number of days per week'] == 'None')]

# melt dataframe
data4_new = data4.melt(id_vars=['Number of days per week'], var_name='Age Group', value_name='Number of kids(\'000)')

# avoid encoding issues
data4_new.replace('5–8','5-8', inplace=True)
data4_new.replace('9–11','9-11', inplace=True)
data4_new.replace('12–14','12-14', inplace=True)
data4_new.replace('15–17','15-17', inplace=True)
data4_new.replace('Total 5–17 years', 'Total 5-17 years', inplace=True)

# for gender female
data4_new['Gender'] = 'Female'
data4_new

data4_new
```

```python
#### for person data
data_both_g = data_both.iloc[12:,:]
data_both_g
```

```python
# same operations applied to all persons data
# make a copy
data5 = data_both_g.copy()
data5.loc['2-7 days'] = data5.iloc[1:5,:].sum()
data5.loc['less than 1 day'] = data5.loc['Total(d)'].sub(data5.loc['2-7 days'])

# reset index
data5.reset_index(inplace=True)

# to avoid encoding issues
# change index values
data5.replace('1–2 days', '1-2 days', inplace=True)
data5.replace('3–4 days', '3-4 days', inplace=True)
data5.replace('5–6 days', '5-6 days', inplace=True)
data5.replace('Total(d)', 'Total', inplace=True)

# filter data
data5 = data5[~(data5['Number of days per week'] == 'Total')]
data5 = data5[~(data5['Number of days per week'] == '2-7 days')]
data5 = data5[~(data5['Number of days per week'] == 'None')]

# melt dataframe
data5_new = data5.melt(id_vars=['Number of days per week'], var_name='Age Group', value_name='Number of kids(\'000)')

# avoid encoding issues
data5_new.replace('5–8','5-8', inplace=True)
data5_new.replace('9–11','9-11', inplace=True)
data5_new.replace('12–14','12-14', inplace=True)
data5_new.replace('15–17','15-17', inplace=True)
data5_new.replace('Total 5–17 years', 'Total 5-17 years', inplace=True)

# for gender female
data5_new['Gender'] = 'Persons'
data5_new

data5_new
```
