```python
# to find tutorial reference
# v1.2
import pandas as pd
import numpy as np
```

```python
# method 2: read_excel
# read multi indexes with $header 
files = 'table 19 average time spent on physical activity, children aged 5-17 years.xls'
sheets = 'Table_19_1'
data = pd.read_excel(files, sheet_name=sheets, skiprows=4, header = [0,1])
```

```python
data
```

```python
data.shape
```

```python
# delete null rows
data1 = data[~data.isnull().all(1)]
data1
```

```python
# drop null columns
data1.dropna(axis= 1, how= 'all', inplace = True)
data1
```

```python
# drop the column index level 0
# don't run yet
data1.columns = data1.columns.droplevel(0)
```

```python
# extract data for age group
data2 = data1.iloc[2:6, :]
data2
```

```python
# reference: https://stackoverflow.com/questions/41221079/rename-multiindex-columns-in-pandas
data2.columns.set_levels(['Transport per day', 'Sports per day', 'Total per day'], level=0, inplace=True)
```

```python
# change column names
data2.rename(columns={'Total active transport(c) ':'Transport total',
                      'Total organised or non-organised(c)': 'Sports total',
                      'Organised': 'Organised sports',
                      'Non-organised': 'Non-organised sports'                      
                      }, inplace=True)
```

```python
data2
```

```python
# reshape the data from a wide format to a long format
data3 = data2.T.stack().reset_index()
```

```python
# Type 1: first level categories
# Type 2: second level categories
new_col = ['Type 1','Type of activity','Age Group','Minutes']
data3.columns = new_col
data3
```

```python
# change values to avoid encoding issues
data3.replace('5–8', '5-8', inplace=True)
data3.replace('9–11', '9-11', inplace=True)
data3.replace('12–14', '12-14', inplace=True)
data3.replace('15–17', '15-17', inplace=True)
```

```python
data3
```

```python
# write results to csv
data3.to_csv('average minutes spent on each activity per day_v1.0.csv', encoding='utf8')
```
